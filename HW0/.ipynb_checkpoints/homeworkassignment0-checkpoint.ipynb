```python
import numpy as np
import csv
```

```python
#N×p array, where N will be the number of samples and p will be the number of attributes per sample in this data set
matrix = np.loadtxt(open("/Users/xpressions11/Downloads/iris.data", "rb"), delimiter=",", usecols=range(4))
```

```python
#here, creating the N-dimensional vector containing each sample’s label/class
labels = [None] * 150
with open('/Users/xpressions11/Downloads/iris.data', 'r') as file:
    reader = csv.reader(file)
    temp = list(reader)
    
for i in range(150):
    labels[i] = temp[i][4]
```

```python
from matplotlib import pylab as plt
sepalLen = matrix[:,0]
sepalWid = matrix[:,1]
petalLen = matrix[:,2]
petalWid = matrix[:,3]
colors = ["r"] * 150
for i in range(50, 100):
    colors[i] = "g"
for i in range(100, 150):
    colors[i] = "b"
```

```python
f, axarr = plt.subplots(4, 4)
f.suptitle('Iris Data (red=setosa, green=versicolor, blue=virginica)', fontsize=15)
f.set_size_inches(15,15)

axarr[0, 0].annotate('Sepal.Length', xy=(0.5, 0.5), xycoords='data', fontsize=8, horizontalalignment='center', verticalalignment='center')
# xycoordsdata uses the coordinate system of the object being annotated (by default)

axarr[0, 1].scatter(sepalWid, sepalLen, c=colors)
axarr[0, 2].scatter(petalLen, sepalLen, c=colors)
axarr[0, 3].scatter(petalWid, sepalLen, c=colors)
axarr[1, 0].scatter(sepalLen, sepalWid, c=colors)
axarr[1, 1].annotate('Sepal.Width', xy=(0.5, 0.5), xycoords='data', fontsize=8, horizontalalignment='center', verticalalignment='center')
axarr[1, 2].scatter(petalLen, sepalWid, c=colors)
axarr[1, 3].scatter(petalWid, sepalWid, c=colors)
axarr[2, 0].scatter(sepalLen, petalLen, c=colors)
axarr[2, 1].scatter(sepalWid, petalLen, c=colors)
axarr[2, 2].annotate('Petal.Length', xy=(0.5, 0.5), xycoords='data', fontsize=8, horizontalalignment='center', verticalalignment='center')
axarr[2, 3].scatter(petalWid, petalLen, c=colors)
axarr[3, 0].scatter(sepalLen, petalWid, c=colors)
axarr[3, 1].scatter(sepalWid, petalWid, c=colors)
axarr[3, 2].scatter(petalLen, petalWid, c=colors)
axarr[3, 3].annotate('Petal.Width', xy=(0.5, 0.5), xycoords='data', fontsize=8, horizontalalignment='center', verticalalignment='center')
plt.savefig("iris.png")
```
