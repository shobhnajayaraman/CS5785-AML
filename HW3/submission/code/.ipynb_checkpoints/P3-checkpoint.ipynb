```python
import numpy as np
np.set_printoptions(threshold=np.inf)

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

# Our 2-dimensional distribution will be over variables X and Y
N = 60
X1 = np.linspace(-3, 3, N)
Y1 = np.linspace(-3, 4, N)

X2 = np.linspace(-3, 3, N)
Y2 = np.linspace(-3, 4, N)

X3 = np.linspace(-3, 3, N)
Y3 = np.linspace(-3, 4, N)


X1, Y1 = np.meshgrid(X1, Y1)
X2, Y2 = np.meshgrid(X2, Y2)
X3, Y3 = np.meshgrid(X3, Y3)

a = .2

# Mean vector and covariance matrix
mu1 = np.array([0.,0.])
Sigma1 = np.array([[ a, 0], [0,  a]])

mu2 = np.array([1.,1.])
Sigma2 = np.array([[ a , 0], [0,  a]])

mu3 = np.array([1.,-1.])
Sigma3 = np.array([[ a , 0], [0,  a]])


# Pack X and Y into a single 3-dimensional array
pos1 = np.empty(X1.shape + (2,))
pos1[:, :, 0] = X1
pos1[:, :, 1] = Y1

pos2 = np.empty(X2.shape + (2,))
pos2[:, :, 0] = X2
pos2[:, :, 1] = Y2

pos3 = np.empty(X3.shape + (2,))
pos3[:, :, 0] = X3
pos3[:, :, 1] = Y3


def multivariate_gaussian(pos1, mu1, Sigma1):
    """Return the multivariate Gaussian distribution on array pos.

    pos is an array constructed by packing the meshed arrays of variables
    x_1, x_2, x_3, ..., x_k into its _last_ dimension.

    """

    n1 = mu1.shape[0]
#     print(n1)
    Sigma_det1 = np.linalg.det(Sigma1)
    Sigma_inv1 = np.linalg.inv(Sigma1)
    N1 = np.sqrt((2*np.pi)**n1 * Sigma_det1)
    # This einsum call calculates (x-mu)T.Sigma-1.(x-mu) in a vectorized
    # way across all the input variables.
    fac1 = np.einsum('...k,kl,...l->...', pos1-mu1, Sigma_inv1, pos1-mu1)

    return np.exp(-fac1 / 2) / N1

# The distribution on the variables X, Y packed into pos.
Z1= multivariate_gaussian(pos1, mu1, Sigma1)
Z2= multivariate_gaussian(pos2, mu2, Sigma2)
Z3= multivariate_gaussian(pos3, mu3, Sigma3)

# print(Z1[30]);
# print(Z1.shape);
# Create a surface plot and projected filled contour plot under it.
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X1, Y1, Z1/(Z1+Z2+Z3), rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap=cm.viridis)

cset = ax.contourf(X1, Y1, Z1, zdir='z', offset=-0.15, cmap=cm.viridis)

# Adjust the limits, ticks and view angle
ax.set_zlim(-0.15,10)
ax.set_zticks(np.linspace(0,0.2,5))
ax.view_init(27, -21)

plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X1, Y1, Z2/(Z1+Z2+Z3), rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap=cm.viridis)

cset = ax.contourf(X1, Y1, Z1, zdir='z', offset=-0.15, cmap=cm.viridis)

# Adjust the limits, ticks and view angle
ax.set_zlim(-0.15,10)
ax.set_zticks(np.linspace(0,0.2,5))
ax.view_init(27, -21)

plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot_surface(X1, Y1, Z3/(Z1+Z2+Z3), rstride=3, cstride=3, linewidth=1, antialiased=True,
                cmap=cm.viridis)

cset = ax.contourf(X1, Y1, Z1, zdir='z', offset=-0.15, cmap=cm.viridis)

# Adjust the limits, ticks and view angle
ax.set_zlim(-0.15,10)
ax.set_zticks(np.linspace(0,0.2,5))
ax.view_init(27, -21)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

with open("Problem 3/P3.txt", 'r') as erruptions:
     features = np.array([line.split(' ') for line in erruptions][1:], dtype=float)[:,1:]
print(features.shape)
# print(features)

plt.scatter(features[:,0],features[:,1])
plt.title("Erruption Data")
plt.xlabel("Duration of Erruption(mins)")
plt.ylabel("Time Since Last Erruption(mins)")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Calculate mean erruption duration
mu_erruption_duration=features[:,0].mean()

# Calculate mean time between erruptions
mu_time_between_erruptions=features[:,1].mean()

# Put the two numbers above on [2,1] array
mu_features = np.array([mu_erruption_duration,mu_time_between_erruptions])

# Create new feature vector with mean subtracted
features_centered = features-mu_features

# Compute the variance across feature 0 (all rows, collumn 0)
var1 = (np.var(features[:,0]))

# Compute the variance across feature 1 (all rows, collumn 1)
var2 = (np.var(features[:,1]))

# Compute standarized feature 0 = Ratio of (i) signed distance from mean and (ii) STANDARD DEVIATION, NOT VARIANCE
erruption_duration_std = features_centered[:,0]/np.sqrt(var1)

###################################################
# Verify variance of standarized feature 0 = 1
# print(np.var(erruption_duration_std))
###################################################

# Compute standarized feature 1 = Ratio of (i) signed distance from mean and (ii) STANDARD DEVIATION, NOT VARIANCE
time_between_erruptions_std = features_centered[:,1]/np.sqrt(var2)

features_std =np.column_stack((erruption_duration_std,time_between_erruptions_std))
# print (features_std)

###################################################
#  Verify variance of standarized feature 1 = 1  ##
#  print(np.var(erruption_duration_std))         ##
###################################################

# Map individual standard vectors to single DISTANCE vector
dist_std = np.zeros(len(erruption_duration_std))
for i,j in enumerate(dist_std):
    dist_std[i]= np.linalg.norm([erruption_duration_std[i],time_between_erruptions_std[i]])

covar_mat_array = [np.array([[1,0]  ,[0,1]]),np.array([[1,0]  ,[0,1]])]
means_array = np.array([[-.5,-1.5],[.7,1.1]])
means_array_update = means_array
pi_hat_case = 0.5

c_case = np.zeros([2,len(features_std)])
d = np.zeros([2,len(features_std),len(features_std)])

def get_responsibilities(features,means, covars,pi_hat,c):    
    rv0 = multivariate_normal(means[0],covars[0])
    rv1 = multivariate_normal(means[1],covars[0])
    for i,j in enumerate(erruption_duration_std):
#         print(i)
        c[0,i]= (1-pi_hat) * rv0.pdf(features_std[i])/((1-pi_hat) * rv0.pdf(features_std[i])+(pi_hat) * rv1.pdf(features_std[i]))
        c[1,i]=   (pi_hat) * rv1.pdf(features_std[i])/((1-pi_hat) * rv0.pdf(features_std[i])+(pi_hat) * rv1.pdf(features_std[i]))

#         PRINTS COVARANCES IF ENABLED
#         print(c[0,i])
#         print(c[1,i])

        d[0,i,i]=c[0,i]
        d[1,i,i]=c[1,i]

    means[0] = np.dot(np.transpose(c[0,:]),features_std)/(np.dot(np.transpose(c[0,:]),np.ones(len(features_std))))
#     print('mean 1 is ', means[0])

    means[1] = np.dot(np.transpose(c[1,:]),features_std)/(np.dot(np.transpose(c[1,:]),np.ones(len(features_std))))
#     print('mean 2 is ', means[1])   

    covars[0] = np.linalg.multi_dot([np.transpose(features_std-means[0]),d[0],features_std-means[0]])/(np.dot(np.transpose(c[0,:]),np.ones(len(features_std))) )  
    covars[1] = np.linalg.multi_dot([np.transpose(features_std-means[1]),d[1],features_std-means[1]])/(np.dot(np.transpose(c[1,:]),np.ones(len(features_std))) )

    return(np.array(means))




iterations = 500
v_mean = np.empty([iterations,2,2])

v_mean[0]=means_array_update

ax = plt.axes()
for _,i in enumerate(range(iterations-1)):
    v_mean[i+1]=get_responsibilities(features_std,means_array,covar_mat_array,pi_hat_case,c_case)

    dx1=v_mean[i+1,0,0]-v_mean[i,0,0]
    dy1=v_mean[i+1,0,1]-v_mean[i,0,1]

    ax.arrow(v_mean[i,0,0], v_mean[i,0,1], dx1, dy1, head_width=.05, head_length=.1, fc='b', ec='b',width=.03)

    dx2=v_mean[i+1,1,0]-v_mean[i,1,0]
    dy2=v_mean[i+1,1,1]-v_mean[i,1,1]

    ax.arrow(v_mean[i,1,0], v_mean[i,1,1], dx2, dy2, head_width=.05, head_length=.01, fc='g', ec='g',width=.03)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.plot(v_mean[iterations-1,0,0], v_mean[iterations-1,0,1], marker='o', markersize=8, color="b")
plt.plot(v_mean[iterations-1,1,0], v_mean[iterations-1,1,1], marker='o', markersize=8, color="g")
plt.scatter(erruption_duration_std,time_between_erruptions_std, color='gray')
plt.title("Erruption Data")
plt.xlabel("Duration of Erruption(Zero Mean - Unit Variance)")
plt.ylabel("Time Since Last Erruption(Zero Mean - Unit Variance)")
plt.show()

print("Centers this is iteration are:")
print("(Data Centered at Zero - Normalized to Unit Variance)")
print(v_mean[iterations-1,0])
print(v_mean[iterations-1,1])
print('\n\n')
```
